```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

Leemos la base de datos csv

Crearemos un nueva columna basada en bad_loans donde lo recategorizaremos . Luego borraremos bad_loans

```python
loans['safe_loans']=loans['bad_loans'].apply(lambda x: 1 if x==0 else -1)
loans.drop('bad_loans',axis=1)
```

```python
loans['safe_loans'].value_counts()
```

Contamos  y analizamos el porcentaje de prestas seguros y riesgosos 

```python
safe_loans=( (loans['safe_loans']==1).sum()/len(loans['safe_loans']) )*100
print ('safe_loans % :',safe_loans)
print ('risky_loans % :',100-safe_loans)
```

Seleccionaremos las características que nos importan más y nuestro target 

```python
features = ['grade',                     # grade of the loan
            'sub_grade',                 # sub-grade of the loan
            'short_emp',                 # one year or less of employment
            'emp_length_num',            # number of years of employment
            'home_ownership',            # home_ownership status: own, mortgage or rent
            'dti',                       # debt to income ratio
            'purpose',                   # the purpose of the loan
            'term',                      # the term of the loan
            'last_delinq_none',          # has borrower had a delinquincy
            'last_major_derog_none',     # has borrower had 90 day or worse rating
            'revol_util',                # percent of available credit being used
            'total_rec_late_fee',        # total late fees received to day
           ]

target = 'safe_loans'                    # prediction target (y) (+1 means safe, -1 is risky)

# Extract the feature columns and target column
loans = loans[features + [target]]
```

A continuación convertiremos las variables categóricas en binarias 

```python
categorical_variables = []
for feat_name, feat_type in zip(loans.columns, loans.dtypes):
    if feat_type == 'object':
        categorical_variables.append(feat_name)
```

```python
loan_categorical = pd.get_dummies(loans, columns=categorical_variables)
```

```python
#cambiamos el tipo de dato booleano por entero
for i in loan_categorical.columns:
    if i not in(loans.columns):
        loan_categorical[i]=loan_categorical[i].astype(int)
del i
```

```python
loans=loan_categorical
del loan_categorical
```

Creamos un pequeño subconjunto de datos ...

```python
validation_safe_loans = validation_data[validation_data[target] == 1]
validation_risky_loans = validation_data[validation_data[target] == -1]

sample_validation_data_risky = validation_risky_loans[0:2]
sample_validation_data_safe = validation_safe_loans[0:2]

sample_validation_data = pd.concat([sample_validation_data_safe,sample_validation_data_risky])
sample_validation_data
```

```python
def accuracy_cal(predicciones,target):
    cont=0
    for i,j in zip(predicciones,target):
        if i==j:
            cont=cont+1
    print (cont/len(predicciones))
```

Precisión del modelo en el pequeño subconjunto de datos

```python
accuracy_cal(decision_tree_model.predict(sample_validation_data.drop('safe_loans',axis=1)),sample_validation_data['safe_loans'])
```

Cuantificando el costo de equivocaciones

```python
F_P=0 #falsos positivos
F_N=0 #falsos negativos
ACT=0 # Acierto
for i,j in zip(decision_tree_model.predict(validation_data.drop('safe_loans',axis=1)),validation_data[target]):
    
    if j==1 and i!=j:
        F_N=F_N+1
    elif j==-1 and i!=j:
        F_P=F_P+1
    else:
        ACT=ACT+1

print ('Costo de Falsos Negativos : ',F_N*10000,' Cantidad : ' , F_N)
print ('Costo de Falsos Positivos : ',F_P*20000,' Cantidad : ' , F_P)
print ('Cantidad de aciertos : ' , ACT)
print('Costo Total : ',F_N*10000 + F_P*20000)
```
